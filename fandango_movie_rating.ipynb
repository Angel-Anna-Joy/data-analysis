```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample movie_ratings_selected data for 2015 and 2016
np.random.seed(0)  # For reproducibility
movie_2015 = np.random.normal(loc=3.5, scale=0.5, size=1000)  # Ratings centered around 3.5
movie_2016 = np.random.normal(loc=3.2, scale=0.6, size=1000)  # Ratings centered around 3.2

# Create the KDE plot
plt.figure(figsize=(6, 4))

# Plot KDE for 2015 and 2016
sns.kdeplot(movie_2015, shade=True, color='blue', label='2015', alpha=0.6)
sns.kdeplot(movie_2016, shade=True, color='green', label='2016', alpha=0.6)

# Customize the plot
plt.title('Kernel Density Estimation of Movie Ratings', fontsize=18)
plt.xlabel('Movie Rating', fontsize=14)
plt.ylabel('Density', fontsize=14)

# Set x-axis range (ratings between 0 and 5)
plt.xlim(0, 5)

# Add a legend
plt.legend(title="Year")

# Show grid for better visibility
plt.grid(False)

# Show the plot
plt.show()
```

```python
mean_2015 = fandango_2015['Fandango_Stars'].mean()
mean_2016 = movie_2016['fandango'].mean()

mode_2015 = fandango_2015['Fandango_Stars'].mode()
mode_2016 = movie_2016['fandango'].mode()

median_2015 = fandango_2015['Fandango_Stars'].median()
median_2016 = movie_2016['fandango'].median()
```
